```python
import pandas as pd
```

```python
df = pd.read_excel(
    io='supermarkt_sales.xlsx',
    engine='openpyxl',
    sheet_name="Sales",
    skiprows=3,
    usecols='B:R',
    nrows=1000
)
df['hour'] = pd.to_datetime(df['Time'],format="%H:%M:%S").dt.hour
df
```

```python
# ['Yangon', 'Naypyitaw', 'Mandalay']  ['Member', 'Normal'] ['Female', 'Male']

city='Yangon'
customer_type='Member'
gender='Female'

df_selection = df.query(
    "City == @city & Customer_type == @customer_type & Gender == @gender"
)

df_selection
```

```python
df_selection.groupby(by=['Product line']).sum()[['Total']].sort_values("Total")
```

```python
df_selection.groupby(by=['hour']).sum()[['Total']]
```
